```python
import os
import django

os.environ.setdefault("DJANGO_SETTINGS_MODULE", "yourproject.settings")
django.setup()
```

```python
from taskapp.models import TaskSection, TaskType2, TaskType1, TaskType2Variant

from pathlib import Path
import json
import pathlib 
import shutil

root = Path("data")
root.mkdir(parents=True, exist_ok=True)

for sec in TaskSection.objects.all():
    print('üîµ')
    print(sec.id)
    print(sec.name)
    print(sec.slug)
    print(sec.image.path if sec.image else '')
    print()
    
    sec_images = root.joinpath(sec.slug)
    print('sec_images: ', sec_images)
    
    allt2 = []
    for task in Task.objects.filter(section=sec):
        t2 = {'id': task.number}
        print('\t', task.number)
        if task.type.slug == 'test':
            q = TaskType2.objects.filter(id=task.id_inclass).first()
            if q:
                t2['type'] = 'multi'
                t2['text'] = q.text
                if q.image:
                    im = pathlib.Path(q.image.path)
                    im2 = sec_images.joinpath('img-' + str(task.number) + str(im.suffix))
                    t2['image'] = str(im2.name)
                    sec_images.mkdir(parents=True, exist_ok=True)
                    shutil.copy(str(im), str(im2))
                    
                    
                t2['explanation'] = q.explanation
                if q.explanation_image:
                    im = pathlib.Path(q.explanation_image.path)
                    im2 = sec_images.joinpath('img-' + str(task.number) + '-exp' + str(im.suffix))
                    t2['explanation_image'] = str(im2.name)
                    sec_images.mkdir(parents=True, exist_ok=True)
                    shutil.copy(str(im), str(im2))

                options = TaskType2Variant.objects.filter(task=q)
                ops2 = [] 
                for opt in options:
                    o2 = {'text': opt.text}
                    if opt == q.variant_true:
                        o2['istrue'] = True
                    ops2.append(o2)
                t2['options'] = ops2 
        else:
            print('üçé ', 'task.number: ', task.number)
            print()
            q = TaskType1.objects.get(id=task.id_inclass)
            t2['type'] = 'text'
            t2['text'] = q.text
            if q.image: 
                im = pathlib.Path(q.image.path)
                im2 = sec_images.joinpath('img-' + str(task.number) + str(im.suffix))
                t2['image'] = str(im2.name)
                sec_images.mkdir(parents=True, exist_ok=True)
                shutil.copy(str(im), str(im2))
                
            t2['answer'] = q.answer
            t2['explanation'] = q.explanation
            if q.explanation_image:
                im = pathlib.Path(q.explanation_image.path)
                im2 = sec_images.joinpath('img-' + str(task.number) + '-exp' + str(im.suffix))
                t2['explanation_image'] = str(im2.name)
                sec_images.mkdir(parents=True, exist_ok=True)
                shutil.copy(str(im), str(im2))

        allt2.append(t2)
    
    print()
    print()
    d2 = allt2
    if d2:
        sec_json = root.joinpath(sec.slug + '.json')
        sec_json.write_text(json.dumps(d2, ensure_ascii=False, indent=4), encoding="utf-8")
    

    
```
